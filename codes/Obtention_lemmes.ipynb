```python
def clean_text(string):
    first = re.sub("\n","",string)
    new_string = re.sub("[0-9]. ",'',first)
    new_string_2 = re.sub('- ','',new_string)
    final = re.sub(' . ',' ', new_string_2)
    final2 = re.sub('-, ','',final)
    final3 = re.sub('[0-9]','',final2)
    return final3
```

```python
df['textes'] = df['Textes'].apply(clean_text)
```
