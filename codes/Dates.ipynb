```python
for i in numéros_livre_1:
    
    
    num = i 

    with open(f'1.{num}.txt','r') as file:
        data = file.readlines()

    liste = []
    for i in range(len(data)-2):
        if data[i].startswith('dat.') and data[i+1] == '\n':
            liste.append(data[i])
    liste_date = []
    for i in liste:
        liste_date = i.split()

    date = []    
    pattern = re.compile(r'^\d{4}-\d{4}|^\d{4}')

    for item in liste_date:
        if pattern.match(item):
            date.append(item)

    date_finale = []



    punctuations = '''!()[]{};:'"\,<>./?@#$%^&*_~'''

    for i in range(len(date)):
        date[i] = date[i].rstrip(punctuations)

    pattern_2 = r'^\d{4}-\d{4}$|^\d{4}$'   

    date = [re.findall(pattern_2, s)[0] for s in date if re.findall(pattern_2, s)]
    print(date)


    if len(date) > 1:

        date = list(set(date))
        date = list(map(int, date))
        lowest = str(min(date))
        highest = str(max(date))
        date = [lowest,highest]
        date = [date[0]+'-'+date[1]]

        dates.extend(date)

    else:
        dates.extend(date)
#print(len(dates))
#print(dates)
```
