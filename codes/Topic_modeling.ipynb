```python
import pandas as pd
import numpy as np
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
path = '/home/noe/Documents/2022-2023/ENC/MÃ©moire/Dataframes/PDV_grams.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df.pop('Unnamed: 0')
```

```python
sample = df[df['Livre']=='Livre 1']
```

```python
sample = sample.reset_index()
```

```python
from gensim.corpora import Dictionary
from gensim.models import TfidfModel
```

```python
dictionary = Dictionary(sample['lemma'].str.split())
dictionary.filter_extremes(no_below=5, no_above=0.5)
corpus = [dictionary.doc2bow(tokens) for tokens in sample['lemma'].str.split()]
tfidf = TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
from gensim.models.ldamodel import LdaModel
lda_model = LdaModel(corpus=corpus_tfidf, id2word=dictionary, num_topics=5, passes=20, alpha='auto')
```

```python
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
from gensim.models import CoherenceModel

# Compute coherence score
coherence_model_lda = CoherenceModel(model=lda_model, texts=sample['lemma'].str.split(), dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```
