```python
texte_int = []

for i in sorted(os.listdir(path)):
    
    img = path + i
    
    ocr_result = pytesseract.image_to_string(img)
    
    sentence = ocr_result.split('\n')[:-3]
    
    texte_int.extend(sentence)
```

```python
liber_primus = texte_int.index('LIBER PRIMUS')
liber_secundus = texte_int.index('LIBER SECUNDUS')
liber_tertius = texte_int.index('LIBER TERTIUS')
liber_quartus = texte_int.index('LIBER QUARTUS')
liber_quintus = texte_int.index('LIBER QUINTUS')
liber_sextus = texte_int.index('LIBER SEXTUS')
liber_septimus = texte_int.index('LIBER SEPTIMUS')
liber_octavus = texte_int.index('LIBER OCTAVUS')
liber_nonus = texte_int.index('LIBER NONUS')
liber_decimus = texte_int.index('LIBER DECIMUS')
```

```python
livre_un = texte_int[liber_primus:liber_secundus]
livre_deux = texte_int[liber_secundus:liber_tertius]
livre_trois = texte_int[liber_tertius:liber_quartus]
livre_quatre = texte_int[liber_quartus:liber_quintus]
livre_cinq = texte_int[liber_quintus:liber_sextus]
livre_six = texte_int[liber_sextus:liber_septimus]
livre_sept = texte_int[liber_septimus:liber_octavus]
livre_huit = texte_int[liber_octavus:liber_nonus]
livre_neuf = texte_int[liber_nonus:liber_decimus]
livre_dix = texte_int[liber_decimus:]
```

```python
def calculate_waiting_time(num_people, movie_length, friend_ratings):
    base_waiting_time = 10
    friend_score = sum(friend_ratings) / len(friend_ratings)
    friend_score_factor = 0.5
    additional_waiting_time = movie_length * friend_score * friend_score_factor
    total_waiting_time = base_waiting_time + additional_waiting_time
    return total_waiting_time

# Example usage
num_people_waiting_for = 2
movie_length_minutes = 115
friend_ratings = [3, 4]
waiting_time = calculate_waiting_time(num_people_waiting_for, movie_length_minutes, friend_ratings)
print("You should wait for", waiting_time, "minutes.")
```
