```python
def generate_skipgrams_indices(indices, gram_length, skip):
    skipgrams = []
    
    for central_index in range(len(indices)):
        for offset in range(-skip, skip+1):
            new_index = central_index + offset
            if 0 <= new_index < len(indices) and offset != 0:
                skipgrams.append([central_index] + [new_index] * (gram_length - 1))
    
    return skipgrams

def skip_grams_indices(string):
    gram_length = 3
    skip = 1
    word_indices = list(range(len(string.split())))
    skipgram_indices = generate_skipgrams_indices(word_indices, gram_length, skip)
    return skipgram_indices
def count_matching_words(tuple1, tuple2):
    return sum(1 for word1 in tuple1 if word1 in tuple2)

def find_matching_tuples(main_tuple, tuple_list):
    matching_tuples = []

    for tuple2 in tuple_list:
        if count_matching_words(main_tuple, tuple2) >= 2:
            matching_tuples.append(main_tuple)

    return matching_tuples
def indices_texte(string):
    value = string.split()
    liste = list(range(0,len(value)))
    
    skipgrammes = list(skipgrams(liste,3,1))
    
    return skipgrammes
def indexes_words(liste):
    flattened_list = [num for tup in liste for num in tup]
    return list(set(flattened_list))
def get_index(liste1,liste2):
    
    liste_finale = []
    
    for i in liste1:
        liste_finale.append(liste2[i])
    return liste_finale
def indices_trigrammes(liste1,liste2):
    liste_index = []
    for i in liste1:
        index = liste2.index(i)
        liste_index.append(index)
        
    return liste_index
def tri_to_pos_2(liste,dictio):
    final = []
    for i in liste:
        inter = []
        for j in i:
            inter.append(dictio.get(j))
        final.append(inter)
    return final
```

```python
def skipgrammes(string):
    gram_length = 3
    skip = 1

    # Tokenize the sentence
    tokens = string.split()

    # Compute skip n-grams
    skip_ngrams = list(skipgrams(tokens, gram_length, skip))

    return skip_ngrams

def common(liste1,liste2):
    final = [trigram for trigram in liste1 if trigram in liste2]
    
    if len(final) == 0:
        return False
    else:
        return final
def delete_tuples(liste):
    final = []
    for element in liste:
        if not element in words_tuples:
            final.append(element)
    if len(final) == 0 or len(final) == 1:
        return False
    else:
        return final
def string_to_list(string):
    return ast.literal_eval(string)

def tri_to_pos(liste):
    final = []
    for i in liste:
        inter = []
        for j in i:
            inter.append(dictionnaire.get(j))
        final.append(inter)
    return final
def keep_pos(x):
    liste_pos = ['ADJqua','VER','ADV','NOMcom']
    final = []
    for sublist in x:
        counter = 0
        for word in sublist:
            if word in liste_pos:
                counter += 1      
        if counter > 1:
            final.append(sublist)
        else:
            final.append(False)
    if all(value == False for value in final):
        return False
    else:
        return final
def un_tri(liste):
    if len(liste) - liste.count(False) == 1:
        return False
    else:
        return liste
def double(liste):
    if len(liste) != 2:
        return liste
    else:
        return liste[0] == liste[1]
def skip_leven(liste1,liste2):

    x = liste1.split()
    y= liste2.split()
    
    pierre_mot = []
    texte_mot = []
    pierre_indice = []
    texte_indice= []

    for idx1, word1 in enumerate(x):
        for idx2, word2 in enumerate(y):
            similarity = textdistance.levenshtein.normalized_similarity(word1, word2)
            if 0.8 <= similarity < 1 and word1 in x and word2 in y:
                pierre_indice.append(idx2)
                texte_indice.append(idx1)
                pierre_mot.append(word2)
                texte_mot.append(word1)
    hen = []
    for i in texte_indice:
        segment = segment_lists(x,i-3,i+3,0)
        hen.append(' '.join(segment))
    bertrand = []
    for i in hen:
        bertrand.append(skipgrammes(i))
    kat = []
    for liste in bertrand:
        tim = []
        tim.append(liste[1:4])
        tim.append(liste[5:8])
        tim.append(liste[9:12])
        kat.append(tim)
    tuples_texte = [ [tuple for sublist in inner_list for tuple in sublist] for inner_list in kat]
    ken = []
    for i in pierre_indice:
        segment = segment_lists(y,i-3,i+3,0)
        ken.append(' '.join(segment))
    alex = []
    for i in ken:
        alex.append(skipgrammes(i))
    eva = []
    for liste in alex:
        ju = []
        ju.append(liste[1:4])
        ju.append(liste[5:8])
        ju.append(liste[9:12])
        eva.append(ju)
    tuples_pierre = [ [tuple for sublist in inner_list for tuple in sublist] for inner_list in eva]
    last = []
    for i in range(len(tuples_texte)):
        yeah = []
        for j in tuples_texte[i]:
            result = find_matching_tuples(j,tuples_pierre[i])
            yeah.append(result)
        last.append(yeah)
    return last
```

```python
def get_segments_lemmes(liste,liste2,dictio):
    liste_pos = ['ADJqua','VER','ADV','NOMcom']
    liste1 = ['**' + word + '**' if dictio.get(word) in liste_pos and word in liste else word for word in liste2]
    liste2 = ['**' + word + '**' if dictio.get(word) in liste_pos and word in liste2 else word for word in liste]
    
    
    
    return liste1,liste2
def include_lemmas(liste1,liste2,dictio1,dictio2):

    yo = []
    for word in liste1:
        if word in list(dictio1.keys()):
            yo.append(dictio1.get(word))
        else:
            yo.append('bertrand')
    ya = []
    for word in liste2:
        if word in list(dictio1.keys()):
            ya.append(dictio1.get(word))
        else:
            ya.append('bertrand')
    un, deux = get_segments_lemmes(yo,ya,dictio2)
    indice = []
    for i in range(len(un)):
        if un[i].startswith('**'):
            indice.append(i)
    deuxdice = []
    for i in range(len(deux)):
        if deux[i].startswith('**'):
            deuxdice.append(i)
    return indice, deuxdice
def levensthtein(liste_1,liste_2):
    pierre_mot = []
    texte_mot = []

    for idx1, word1 in enumerate(liste_1):
        for idx2, word2 in enumerate(liste_2):
            similarity = textdistance.levenshtein.normalized_similarity(word1, word2)
            if 0.8 <= similarity < 1 and word1 in liste_1 and word2 in liste_2:
                pierre_mot.append(word2)
                texte_mot.append(word1)
    for i in range(len(liste_1)):
        if liste_1[i] in texte_mot:
            liste_1[i] = f'**{liste_1[i]}**'
    for i in range(len(liste_2)):
        if liste_2[i] in pierre_mot:
            liste_2[i] = f'**{liste_2[i]}**'
    return liste_1, liste_2
```

```python
def segment_lemme(viard,pierre,index1,index2,diction,pos_lemma):
    
    liste_pos = ['ADJqua','VER','ADV','NOMcom']
    
    liste_fin = []

    for i in range(len(index1)):

        premier_indice = min(index1[i])
        dernier_indice = max(index1[i])
        debut_fenetre = max(0, premier_indice - 10)
        fin_fenetre = min(dernier_indice + 10, len(viard.split()))
        debut_fenetre

        resultat = viard.split()[debut_fenetre:fin_fenetre]



        premier_indice_2 = min(index2[i])
        dernier_indice_2 = max(index2[i])
        debut_fenetre_2 = max(0, premier_indice_2 - 10)
        fin_fenetre_2 = min(dernier_indice_2 + 10, len(pierre.split()))
        resultat_2 = pierre.split()[debut_fenetre_2:fin_fenetre_2]

        new_result = []
        for i in resultat:
            new_result.append(diction.get(i))

        new_result_2 = []
        for i in resultat_2:
            new_result_2.append(diction.get(i))

        counter = 0
        for i in new_result:
            if i in new_result_2 and pos_lemma.get(i) in liste_pos:
                counter += 1
        if counter > 3:
            liste_fin.append(True)
        else:
            liste_fin.append(False)
            
    return liste_fin
def toutes_fausses(liste):
    for element in liste:
        if element:
            return False

def create_lemmes(boolean,indices_viard,indices_pdv,viard,pierre,diction):
    
    liste_pos = ['ADJqua','VER','ADV','NOMcom']
    
    signe_viard = []
    signe_pdv = []

    for i in range(len(boolean)):
        if boolean[i] == True:
            signe_viard.append(indices_viard[i])
            signe_pdv.append(indices_pdv[i])
    
    liste_fin_viard = []
    liste_fin_pdv = []
    liste_texte_viard = []
    liste_texte_pdv = []

    for i in range(len(signe_viard)):

        premier_indice = min(signe_viard[i])
        dernier_indice = max(signe_viard[i])
        debut_fenetre = max(0, premier_indice - 10)
        fin_fenetre = min(dernier_indice + 10, len(viard.split()))

        resultat = viard.split()[debut_fenetre:fin_fenetre]



        premier_indice_2 = min(signe_pdv[i])
        dernier_indice_2 = max(signe_pdv[i])
        debut_fenetre_2 = max(0, premier_indice_2 - 10)
        fin_fenetre_2 = min(dernier_indice_2 + 10, len(pierre.split()))
        resultat_2 = pierre.split()[debut_fenetre_2:fin_fenetre_2]

        new_result = []
        for i in resultat:
            new_result.append(diction.get(i))

        new_result_2 = []
        for i in resultat_2:
            new_result_2.append(diction.get(i))
                
        
        liste_fin_viard.append(new_result)
        liste_fin_pdv.append(new_result_2)
        liste_texte_viard.append(resultat)
        liste_texte_pdv.append(resultat_2)
            
    return liste_fin_viard,liste_fin_pdv,liste_texte_viard,liste_texte_pdv

def mise_en_gras(liste1,liste2):
    
    final_un = []
    final_deux = []
    
    for i in range(len(liste1)):
        
        one = ['**' + word + '**' if  word in liste1[i] else word for word in liste2[i]]
        two = ['**' + word + '**' if  word in liste2[i] else word for word in liste1[i]]
        
        final_un.append(one)
        final_deux.append(two)
        
    return final_deux,final_un

def texte_gras(liste1,liste2):
    
    aya = []
    for i in range(len(liste1)):
        inter = []
        for j in range(len(liste1[i])):
            if not liste1[i][j].startswith('**'):
                inter.append(liste2[i][j])
            else:
                inter.append(f'**{liste2[i][j]}**')
        aya.append(inter)
    return aya
```

```python
def skip_grams(string):
    gram_length = 3
    skip = 1
    liste = string.split()
    final = list(skipgrams(liste, gram_length, skip))
    
    return final
def common(liste1,liste2):
    final = [trigram for trigram in liste1 if trigram in liste2]
    
    if len(final) == 0:
        return False
    else:
        return final
def delete_tuples(liste):
    final = []
    for element in liste:
        if not element in words_tuples:
            final.append(element)
    if len(final) == 0 or len(final) == 1:
        return False
    else:
        return final
def string_to_list(string):
    return ast.literal_eval(string)

def tri_to_pos(liste):
    final = []
    for i in liste:
        inter = []
        for j in i:
            inter.append(dictionnaire.get(j))
        final.append(inter)
    return final
def keep_pos(x):
    liste_pos = ['ADJqua','VER','ADV','NOMcom']
    final = []
    for sublist in x:
        counter = 0
        for word in sublist:
            if word in liste_pos:
                counter += 1      
        if counter > 1:
            final.append(sublist)
        else:
            final.append(False)
    if all(value == False for value in final):
        return False
    else:
        return final
def un_tri(liste):
    if len(liste) - liste.count(False) == 1:
        return False
    else:
        return liste
def double(liste):
    if len(liste) != 2:
        return liste
    else:
        return liste[0] == liste[1]
```
