```python
def topic_list(text):
    text = text.split(';')
    for i in range(len(text)):
        index = text[i].index('(')
        text[i] = text[i][:index]
        text[i] = text[i].replace(' ','')
    for i in range(len(text)):
        if ',' in text[i]:
            split_values = text[i].split(',')
            text.pop(i)
            for j in range(len(split_values)):
                text.insert(i+j, split_values[j])
    text = sorted(list(set(text)))
    return text
```
