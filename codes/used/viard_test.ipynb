```python
def skip_grams(string):
    gram_length = 3
    skip = 1
    liste = string.split()
    final = list(skipgrams(liste, gram_length, skip))
    
    return final
def common(liste1,liste2):
    final = [trigram for trigram in liste1 if trigram in liste2]
    
    if len(final) == 0:
        return False
    else:
        return final
def delete_tuples(liste):
    final = []
    for element in liste:
        if not element in words_tuples:
            final.append(element)
    if len(final) == 0 or len(final) == 1:
        return False
    else:
        return final
def string_to_list(string):
    return ast.literal_eval(string)

def tri_to_pos(liste):
    final = []
    for i in liste:
        inter = []
        for j in i:
            inter.append(dictionnaire.get(j))
        final.append(inter)
    return final
def keep_pos(x):
    liste_pos = ['ADJqua','VER','ADV','NOMcom']
    final = []
    for sublist in x:
        counter = 0
        for word in sublist:
            if word in liste_pos:
                counter += 1      
        if counter > 1:
            final.append(sublist)
        else:
            final.append(False)
    if all(value == False for value in final):
        return False
    else:
        return final
def un_tri(liste):
    if len(liste) - liste.count(False) == 1:
        return False
    else:
        return liste
def double(liste):
    if len(liste) != 2:
        return liste
    else:
        return liste[0] == liste[1]
```
