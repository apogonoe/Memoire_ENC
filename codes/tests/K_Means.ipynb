```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
file = '/home/noe/Documents/2022-2023/ENC/Mémoire/Dataframes/PDV_grams.xlsx'
```

```python
df = pd.read_excel(file)
```

```python
df = df[['Livre','Numéro de la charte','Titre','Date','textes','lemma']]
```

```python
df
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(df['lemma'])
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(tfidf)
```

```python
df['cluster'] = kmeans.labels_
```

```python
df[df['Livre']=='Livre 1']
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
scatter_plot_points = pca.fit_transform(tfidf.toarray())
colors = ["r", "b", "y"]
df['color'] = df['cluster'].map(lambda p: colors[p])

plt.scatter(scatter_plot_points[:, 0], scatter_plot_points[:, 1], c=df['color'])
plt.show()
```
