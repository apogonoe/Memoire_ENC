```python
def clean_text(text):
    char  = [';',':','!','«','»',',','“','”']
    pattern = '[' + ''.join(char) + ']'
    new_str = re.sub(pattern,'', text)
    clean_words = [word for word in new_str.split() if word.lower() not in stopwords_latin]
    new_str = ' '.join(clean_words)
    new_str = re.sub(r'\? [A-Z]','.',new_str)
    new_str = re.sub(r'\?','',new_str)
    pattern = r"\. ([a-z])"
    new_str = re.sub(pattern, r" \1", new_str)
    string = new_str.split('.')
    string = [x for x in string if x != '']
    string = [x for x in string if x != ' ']
    string = [x.lower() for x in string]
    return string
```

```python
def nett_guerin(string):
    string = re.sub('v','u',string)
    string = re.sub('j','i',string)
    return string

df_guerin['text'] = df_guerin['text'].apply(nett_guerin)
guerin = df_guerin['text']
```

```python
pdv = df_PDV['textes']
pdv = pdv.apply(clean_text).to_frame()
```

```python
df = guerin.apply(clean_text).to_frame()
```

```python
for i in range(0,len(pdv)):
    df[f'pdv_{i}'] = pd.Series(str(pdv['textes'][i]),index=df.index)
    df[f'pdv_{i}'] = df[f'pdv_{i}'].apply(ast.literal_eval)  
```

```python
def comparaison(list1, list2):
    result_dict = {}

    for elem1 in list1:
        matching_sentences = []
        for elem2 in list2:
            words1 = elem1.split()
            words2 = elem2.split()
            if sum([1 for word in words1 if word in words2]) >= 2:
                matching_sentences.append(elem2)
        result_dict[elem1] = matching_sentences

    return result_dict
```

```python
for i in tqdm(range(len(df))):
    for j in range(len(pdv)):
        df[f'pdv_{j}'][i] = comparaison(df['text'][i],df[f'pdv_{j}'][i])
        df[f'pdv_{j}'][i] = {k: v for k, v in df[f'pdv_{j}'][i].items() if v}
        df[f'pdv_{j}'][i] = {f'pdv_{j}' : df[f'pdv_{j}'][i]}
```

```python
df_dico = df.loc[:, df.columns != 'text']
```

```python
final = {}
for i in range(len(df_dico)):   
    row = df_dico.iloc[i]
    pdv = {}
    for d in row:
        pdv.update(d)
        pdv = {k: v for k, v in pdv.items() if v}
    final[f'pdv_{i}'] = pdv
```
